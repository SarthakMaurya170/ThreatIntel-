```python
import pandas as pd
```

```python
print("Loading raw data...")
df = pd.read_csv('raw_attack_logs.csv')
```

```python
columns_to_keep = ['datetime', 'host', 'src', 'proto', 'type', 'dpt', 'cc']
df = df[columns_to_keep]
```

```python
df = df.rename(columns={
    'datetime': 'attack_time',
    'host': 'honeypot_host',
    'src': 'attacker_ip',
    'proto': 'protocol',
    'type': 'attack_type',
    'dpt': 'target_port',
    'cc': 'country_code'
})
```

```python
df = df.dropna(subset=['attacker_ip', 'target_port'])
```

```python
output_filename = 'cleaned_attack_logs.csv'
df.to_csv(output_filename, index=False)
print(f"Success! Cleaned {len(df)} rows and saved to {output_filename}. Ready for SQL.")
```
